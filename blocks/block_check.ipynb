```python
import json
import re


def load_js_json_array(path):
    """Extract and load the JSON array from a JS file like 'window.stimuli = [...]'."""
    with open(path, "r", encoding="utf-8") as f:
        content = f.read()
    # Find the array assignment
    match = re.search(r"=\s*(\[.*\]);", content, re.DOTALL)
    if not match:
        raise ValueError("Could not find JSON array in JS file.")
    array_str = match.group(1)
    return json.loads(array_str)


# Usage:
# stimuli = load_js_json_array("blocks/base_range1.js")
```
